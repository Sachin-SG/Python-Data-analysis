```python
import statistics
mean = statistics.mean(df["Age"])
median = statistics.median(df["Age"])
mode =  statistics.mode(df["Age"])
print(mean)
print(median)
print(mode)
```

```python
# Since median is less than mean , the data might be right skewed and have outliers. lets plot the same
x_axis = df["Age"]
y = x_axis.value_counts()
x = x_axis.unique()
yy = y.sort_index()
print(x)
```

```python
# Scatter plot
import matplotlib.pyplot as plt
print(plt.scatter(x,yy))
```

```python
# PLotting the Normal Distribution of a data 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics

x_axis = df["Age"]

# Calculating mean and standard deviation
mean = statistics.mean(x_axis)
sd = statistics.stdev(x_axis)
print("mean", mean)
print("sd",sd)

plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
plt.show()
```

```python
# Plotting Histogram
plt.hist(x_axis,bins = 33)
```

```python
#Data has ouliers in the upper quartile range, so lets remove the values which are in the upper quartile range
def upper_quartile(l,n):
    n = n * 0.75
    if (n.is_integer()):
        uq_value = (l[n]+l[n+1])/2
    else:
        uq_value = l[n+1]
        
    return uq_value

n = len(df["Age"])
uq_val = upper_quartile(df["Age"],n)
print("Upper Quartile :",uq_val )

df = df[df["Age"] <= uq_val]
df
```
