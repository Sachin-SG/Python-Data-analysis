```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from unicodedata import normalize


df_Batting = pd.DataFrame()
df_FOW = pd.DataFrame()
df_Bowling = pd.DataFrame()
table_MN = pd.read_html('https://www.t20worldcup.com/match/100439')
l = len(table_MN)
for i in range(0,l,3):
    df_bat = pd.DataFrame(table_MN[i])
    df_Batting = pd.concat([df_Batting,df_bat],ignore_index=True)
    df_fow = pd.DataFrame(table_MN[i+1])
    df_FOW = pd.concat([df_FOW,df_fow],ignore_index=True)
    df_bowl = pd.DataFrame(table_MN[i+2])
    df_Bowling = pd.concat([df_Bowling,df_bowl],ignore_index=True)
    
print(df_Batting)
print(df_FOW)
print(df_Bowling)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from unicodedata import normalize


df_Batting = pd.DataFrame()
df_FOW = pd.DataFrame()
df_Bowling = pd.DataFrame()
for j in range(37,66):
    url = 'https://www.t20worldcup.com/match/1004' + str(j)
    try:
        table_MN = pd.read_html(url)
    except ValueError:
        continue
    l = len(table_MN)
    for i in range(0,l,3):
        df_bat = pd.DataFrame(table_MN[i])
        df_Batting = pd.concat([df_Batting,df_bat],ignore_index=True)
        df_fow = pd.DataFrame(table_MN[i+1])
        df_FOW = pd.concat([df_FOW,df_fow],ignore_index=True)
        df_bowl = pd.DataFrame(table_MN[i+2])
        df_Bowling = pd.concat([df_Bowling,df_bowl],ignore_index=True)
    
print(df_Batting)
print(df_FOW)
print(df_Bowling)



```
